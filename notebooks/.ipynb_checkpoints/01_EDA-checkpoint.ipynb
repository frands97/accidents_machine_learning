5. Generamos un DataFrame agrupado para extraer información basándonos en cada accidente.

```python
df_grp = df.groupby(by=['C_YEAR', 'C_MNTH', 'C_WDAY', 'C_HOUR', 'C_SEV', 'C_VEHS',
       'C_CONF', 'C_RCFG', 'C_WTHR', 'C_RSUR', 'C_RALN', 'C_TRAF', 'V_ID',
       'V_TYPE', 'V_YEAR'])
```

```python
df_grp.get_group((1999, '01', '1', '00', 2, '09', '01', '01', '3', '5', '1', '18', '07', '06', '1994'))
```

6. A continuación, creamos un bucle para extraer la información que nos interesa de cada vehículo involucrado en los accidentes, desglosándolo de la siguiente manera:


- Coche/Peatones
- Numero Personas
- Edad del conductor
- Fallecidos en Coche? Si/No (se puede decir numero de personas fallecidas)
- Sexo del Conductor
- Habia copiloto
- Habia personas atras?
- Todos los ocupantes usaban seguridad (cinturon)
- Condiciones del Trafico
- Wday
- Hora
- Mes
- Configuración

```python
vehicles_in_accident = []

for lab_comb, dfs in df_grp:
    dic = {}

    mask_conductor = dfs.P_USER == '1'
    cond_conductor = mask_conductor.any()

    if cond_conductor:
        cond_algun_fallecido = (dfs['P_ISEV'] == '3').sum() > 0
        if cond_algun_fallecido:
            dic['fallecidos'] = 1
        else:
            dic['fallecidos'] = 0
        dic['n_personas'] = dfs.shape[0]
        dic['edad'] = dfs.loc[mask_conductor, 'P_AGE'].values[0]
        dic['sexo'] = dfs.loc[mask_conductor, 'P_SEX'].values[0]
        dic['todos_con_seguridad'] = (dfs['P_SAFE'] == '01').all()
        valores_tiempo = dfs.iloc[0, 1:4]
        valores_trafico = dfs.iloc[0, 5:12]
        valores_coche = dfs.iloc[0, 13:15]
        dic.update(valores_tiempo.to_dict())
        dic.update(valores_trafico.to_dict())
        dic.update(valores_coche.to_dict())
        
        vehicles_in_accident.append(dic)
```

7. Y ahora, creamos el DataFrame con los datos normalizados. Como se aprecia, tenemos datos UUs (datos no conocidos), XX (datos que, según la jurisdicción, no se pueden proporcionar), QQs (la elección es distinta a los valores anteriores) y NNs (el elemento de datos no es aplicable). Lo que haremos a continuación será convertirlos en datos NaN y guardarlos en CSV.

```python
df = pd.DataFrame(vehicles_in_accident)
```

```python
df.sample(10)
```

### **Conversión de los valores no identificables a valores NaN**

```python
df.replace(to_replace=['Q+', 'U+', 'X+', 'N+'], value=np.nan, regex=True, inplace=True)
```

### **Histograma**

```python
def plot_feature(col_name, isNumeric): 
    """
    Visualizamos una variable con y sin facetas sobre los accidentes en sí - no. 
     - col_name es el nombre de la variable en el marco de datos
     - full_name es el nombre completo de la variable
     - numeric es verdadero si la variable es numérica, falso si no lo fuera 
    """

    f, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12,3), dpi=90)

    # Plot without 'Fallecidos'
    if isNumeric:
        sns.distplot(df_final.loc[df_final[col_name].notnull(), col_name], kde=False, ax=ax1)
    else:
        sns.countplot(df_final[col_name], order=sorted(df_final[col_name].unique()), color='#5975A4', saturation=1, ax=ax1)
    ax1.set_xlabel(col_name) 
    ax1.set_ylabel('Count') 
    ax1.set_title(col_name) 

    # Plot with 'Fallecidos'
    if isNumeric:
        sns.kdeplot(x=df_final[col_name].loc[df_final['fallecidos']== 1],
           shade= True, label='fallecidos')
        sns.kdeplot(df_final[col_name].loc[df_final['fallecidos']== 0],
            shade= True, label='no fallecidos')
        plt.legend()
        ax2.set_ylabel('')
        ax2.set_title(col_name + ' por fallecimientos')
    
    else:
        plt_data = df_final.groupby(col_name)['fallecidos'].value_counts(normalize=True).to_frame('proportion').reset_index() 
        sns.barplot(x=col_name, y='proportion', hue="fallecidos", data=plt_data, saturation=1, ax=ax2) 
        ax2.set_ylabel('fallecimientos')
        ax2.set_title('fallecidos')
        ax2.set_xlabel(col_name) 
    plt.tight_layout()
```

```python
df_cat = df_final.columns
df_cat
```

```python
for col in df_cat:
    plot_feature(col, False);
```

```python
df_num = df_final[['V_YEAR', 'edad']]

for col in df_num:
    plot_feature(col, True);
```

Como podemos apreciar en las gráficas, lo primero que podemos apreciar es que el grueso de las edades donde se cometen más accidentes de coche están comprendidas entre los 20 y 40 años. En cuanto a las gráficas superiores, se observan que los datos contabilizados de fallecidos se concentran en los años 1980 al 2000. Tiene sentido ya que son los datos cogidos de la muestra con la que estamos trabajando. Contrastemos ahora con otras representaciones gráficas.
